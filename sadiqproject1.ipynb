```python
import pandas as pd
df=pd.read_csv(r"C:\Data science(sadiq)\pre-owned cars.csv")
df.describe()
```

```python
for i in df:
    j=i.replace(' ','_')
    df=df.rename(columns={i:j.capitalize()})
    df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python

columns_to_fill = ['Overall_cost','Reg_year','Reg_number','Engine_capacity(cc)',
                   'Title','Spare_key','Has_insurance','Ownership','Km_driven',
                   'Fuel_type','Make_year','Transmission','Model','Brand'] 

for column in columns_to_fill:
    df[column].fillna(df[column].mode()[0], inplace=True)

print(df[columns_to_fill].isnull().sum())
```

#### Identifying the most preferred car brand in India

```python
#1.What is the most frequent car brand?

most_frequent_brand = df['Brand'].value_counts().idxmax()
print(most_frequent_brand)
```

```python
import matplotlib.pyplot as plt

brand_counts = df['Brand'].value_counts()

plt.figure(figsize=(10, 6))
brand_counts.plot(kind='bar', color='brown')
plt.title('Frequency of Car Brands in the Dataset', fontsize=16)
plt.xlabel('Brand', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#CONCLUSION

#The most frequent car brand in the dataset is Maruti
```

```python
#2.Which is the  most common car models i

most_common_models = df['Model'].value_counts().head(10)  # Get the top 10 most common models
print(most_common_models)


```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get the top 10 most common car models
most_common_models = df['Model'].value_counts().head(10)

# Plot the data
plt.figure(figsize=(10, 6))
sns.barplot(x=most_common_models.values,y=most_common_models.index,palette='magma')
plt.title('Top 10 Most Common Car Models', fontsize=16)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Model', fontsize=12)
plt.gca().invert_yaxis()  # Invert y-axis to have the most common model at the top
plt.tight_layout()
plt.show()
```

```python
# conclusion 

#The most common car model in the dataset is Baleno DELTA PETROL 1.2.
```

```python
#3. What percentage of cars have manual transmission vs automatic transmission?

transmission_counts = df['Transmission'].value_counts(normalize=True) * 100
print(transmission_counts)
```

```python
#5.Which car model has the highest km driven, and what are its details? 

import pandas as pd 

car_highest_km = df.loc[df['Km_driven'].idxmax()]

car_highest_km_details = {
    'Brand': car_highest_km['Brand'],
    'Model': car_highest_km['Model'],
    'Transmission': car_highest_km['Transmission'],
    'Make Year': car_highest_km['Make_year'],
    'Fuel Type': car_highest_km['Fuel_type'],
    'Engine Capacity (CC)': car_highest_km['Engine_capacity(cc)'],
    'Kilometers Driven': car_highest_km['Km_driven'],
    'Ownership': car_highest_km['Ownership'],
    'Price': car_highest_km['Price'],
    'Overall Cost': car_highest_km['Overall_cost'],
    'Has Insurance': car_highest_km['Has_insurance'],
    'Spare Key': car_highest_km['Spare_key'],
    'Registration Number': car_highest_km['Reg_number']
}

car_highest_km_details
```

```python
import matplotlib.pyplot as plt

car_highest_km_details = {
    'Brand': car_highest_km['Brand'],
    'Model': car_highest_km['Model'],
    'Kilometers Driven': car_highest_km['Km_driven'],
    'Price': car_highest_km['Price'],
    'Overall Cost': car_highest_km['Overall_cost'],
    'Engine Capacity (CC)': car_highest_km['Engine_capacity(cc)'],
}
labels = ['Kilometers Driven', 'Price', 'Overall Cost', 'Engine Capacity (CC)']
values = [car_highest_km_details['Kilometers Driven'], car_highest_km_details['Price'], 
          car_highest_km_details['Overall Cost'], car_highest_km_details['Engine Capacity (CC)']]
plt.figure(figsize=(8, 5))
plt.bar(labels, values, color=['grey', 'brown', 'black', 'grey'])
plt.title(f"Car with Highest Kilometers Driven: {car_highest_km_details['Brand']} {car_highest_km_details['Model']}")
plt.ylabel("Values")
plt.show()
```

```python
#Conclusion

#The highest km driven car model is Honda 'Brio S(O) MT'.
```

```python
#6.What is the average km driven by cars of each brand? 

import pandas as pd 

average_km_by_brand = df.groupby('Brand')['Km_driven'].mean().reset_index()
average_km_by_brand.columns = ['Brand', 'Average Kilometers Driven']
average_km_by_brand = average_km_by_brand.sort_values(by='Average Kilometers Driven', ascending=False)
average_km_by_brand
```

```python
import matplotlib.pyplot as plt

average_km_by_brand = df.groupby('Brand')['Km_driven'].mean().reset_index()
average_km_by_brand.columns = ['Brand', 'Average Kilometers Driven']
average_km_by_brand = average_km_by_brand.sort_values(by='Average Kilometers Driven', ascending=False)

plt.figure(figsize=(10, 6))
plt.barh(average_km_by_brand['Brand'], average_km_by_brand['Average Kilometers Driven'], color='grey')
plt.xlabel('Average Kilometers Driven')
plt.ylabel('Brand')
plt.title('Average Kilometers Driven by Car Brand')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

```

```python
# Conclusion 

# The average km driven by cars of each brand is over 30000KM.
```

```python
#7. Which car model has the highest overall cost (price + overall maintenance cost)?
df['Total_cost'] = df['Price'] + df['Overall_cost']
car_highest_total_cost = df.loc[df['Total_cost'].idxmax()]

car_highest_total_cost_details = {
    'Brand': car_highest_total_cost['Brand'],
    'Model': car_highest_total_cost['Model'],
    'Transmission': car_highest_total_cost['Transmission'],
    'Make Year': car_highest_total_cost['Make_year'],
    'Fuel Type': car_highest_total_cost['Fuel_type'],
    'Engine Capacity (CC)': car_highest_total_cost['Engine_capacity(cc)'],
    'Kilometers Driven': car_highest_total_cost['Km_driven'],
    'Ownership': car_highest_total_cost['Ownership'],
    'Price': car_highest_total_cost['Price'],
    'Overall Maintenance Cost': car_highest_total_cost['Overall_cost'],
    'Total Cost (Price + Maintenance)': car_highest_total_cost['Total_cost'],
    'Has Insurance': car_highest_total_cost['Has_insurance'],
    'Spare Key': car_highest_total_cost['Spare_key'],
    'Registration Number': car_highest_total_cost['Reg_number']
}

car_highest_total_cost_details
```

```python
import matplotlib.pyplot as plt

df['Total_cost'] = df['Price'] + df['Overall_cost']

car_highest_total_cost = df.loc[df['Total_cost'].idxmax()]

labels = ['Price', 'Overall Maintenance Cost', 'Total Cost']
values = [car_highest_total_cost['Price'], car_highest_total_cost['Overall_cost'], car_highest_total_cost['Total_cost']]
plt.figure(figsize=(8, 5))
plt.bar(labels, values, color=['grey', 'lightgreen', 'brown'])
plt.title(f"Car with Highest Overall Cost: {car_highest_total_cost['Brand']} {car_highest_total_cost['Model']}")
plt.ylabel("Cost (in currency units)")
plt.tight_layout()
plt.show()
```

```python
#CONCLUSION 

##Based on the dataset, the highest overall cost (price + overall maintenance cost) car is Maruthi 'Baleno DELTA PETROL 1.2'.
```
